```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
import pandas as pd
import numpy as np
```

```python
spark
```

```python
df = spark.read.option("delimiter", ";").csv("bank-additional-full.csv", header=True, inferSchema = True)
```

```python
df.dtypes
```

```python
df = df.cache()
```

```python
data = df.toPandas()
```

```python
data.head()
```

```python
data.info()
```

```python
import warnings

warnings.filterwarnings("ignore")

from sdv.tabular import CTGAN

model = CTGAN(cuda = True, verbose = True)
model.fit(data)
```

```python
syn_data = model.sample(41188)
syn_data.head(5)
```

```python
model.save('demo.pkl')
```

```python
from table_evaluator import load_data, TableEvaluator

print(data.shape, syn_data.shape)
table_evaluator =  TableEvaluator(data, syn_data)

table_evaluator.visual_evaluation()
```

```python
from sdv.evaluation import evaluate

evaluate(syn_data, data, metrics=['CSTest', 'KSTest','KSTestExtended','ContinuousKLDivergence','DiscreteKLDivergence'], aggregate=False)
```

```python
from sdv.metrics.tabular import LogisticDetection

LogisticDetection.compute(data, syn_data)
```

```python
from sdv.metrics.tabular import BinaryLogisticRegression

BinaryLogisticRegression.compute(data, syn_data, target='y')
```

```python
train = data.sample(int(len(data) * 0.75))
test = data[~data.index.isin(train.index)]

BinaryLogisticRegression.compute(test, train, target='y')
```

```python
train = syn_data.sample(int(len(syn_data) * 0.75))
test = syn_data[~syn_data.index.isin(train.index)]

BinaryLogisticRegression.compute(test, train, target='y')
```

```python
from sdv.metrics.tabular import CategoricalCAP

CategoricalCAP.compute(data,syn_data,key_fields=['marital','education'],sensitive_fields=['job'])
```
